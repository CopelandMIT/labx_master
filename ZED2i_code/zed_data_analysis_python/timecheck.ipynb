Compare Timestamps (chrony vs internal clock)

```python
import json
import numpy as np

# File path to the JSON data
file_path = "/usr/local/zed/samples/body tracking/export/JSON export/cpp/build/chrony_test_v1.json"

# Load the JSON data from the file
with open(file_path, 'r') as file:
    data = json.load(file)

# Initialize a list to store time differences
time_differences = []

# Iterate through the data and calculate timestamp differences
for key, value in data.items():
    # Extract timestamps
    chrony_time_us = value.get("chrony_utc_time_microseconds")
    zed_time_ms = value.get("zed_timestamp_milliseconds")
    
    # Check if both timestamps are present
    if chrony_time_us is not None and zed_time_ms is not None:
        # Convert ZED timestamp to microseconds
        zed_time_us = zed_time_ms * 1000
        
        # Calculate the time difference in microseconds
        time_difference_us = abs(chrony_time_us - zed_time_us)
        
        # Store the difference in the list
        time_differences.append(time_difference_us)
        
        # Print results for this entry (optional)
        print(f"Timestamp {key}:")
        print(f"  Chrony Time (us): {chrony_time_us}")
        print(f"  ZED Time (us): {zed_time_us}")
        print(f"  Difference (us): {time_difference_us}\n")
    else:
        print(f"Timestamp {key}: Missing Chrony or ZED timestamp data.\n")

# Convert time differences to a NumPy array for easy calculation
time_differences = np.array(time_differences)

# Calculate mean, median, and standard deviation
mean_diff = np.mean(time_differences)
median_diff = np.median(time_differences)
std_diff = np.std(time_differences)

# Print overall statistics
print("Overall Statistics:")
print(f"  Mean Difference (us): {mean_diff}")
print(f"  Median Difference (us): {median_diff}")
print(f"  Standard Deviation (us): {std_diff}")
```
